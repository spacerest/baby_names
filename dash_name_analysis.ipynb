```python
from jupyter_dash import JupyterDash
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
from dash import dcc
from dash import html
from dash.dependencies import Input, Output
```

```python
df = pd.read_pickle("census_name_percents.pkl")
```

```python
FAVORITES = ['alta', 'lily', 'mirabelle', 'lyra', 
             'serena', 'ella', 'annika', 'edna', 
             'agata', 'ava', 'adie', 'daria', 'sofia', 
             'eleanore', 'lina', 'odette', 'tara', 'yuna', 
             'zoe', 'marina', 'aurora', 'cora', 'amira', 
             'juno', 'vienna', 'elina', 'lana', 'sona']
VOWELS = list("aeiouy")
```

```python
first_letters = {x[0] for x in FAVORITES}
```

```python
last_letters = ['a', 'e', 'o', 'y', 't', 'n', 'r']
```

```python
# what is the proportion of vowels to consonants in the names we like?

favorites_letter_ratios = [(len([c for c in name if c in VOWELS]) / len(name)) 
                           for name in FAVORITES]

x = zip(favorites_letter_ratios, FAVORITES)

# for y in x:
#     print(y)
```

```python
def get_vowel_ratio(name):
    return len([c for c in name if c in VOWELS]) / len(name)
```

```python
# what do these favs have in common? =hardcoded filters on dataframe

df_filtered = df[(df['name'].str.contains(f'^[{"".join(first_letters)}]'))
                &
                (df['sex'] == 'F')
                &
                (df['name'].str.contains(f'[{"".join(last_letters)}]$'))
                
                &
                (df['name'].apply(get_vowel_ratio) >= 0.5) 
                 
                #todo maybe change this into a dash input
                &
                (df['name'].str.len() < 6)
                & 
                (df['name'].str.len() > 2)
                 
                
                ]
# .str.contains('|'.join(includeKeywords)
```

```python
# add a column that has a number for upper third of its own 
# peak popularity value (or TF might work too)

# when is percent_of_births the highest for a name

# df_filtered["in_top_third_of_percent_range"] = df[df_filtered.percent_of_births.max()] 
#     / 3 > 0.33]

# df_filtered[df_filtered['in_top_third_of_percent_range']==df_filtered.groupby('name')['percent_of_births'].transform('max')]


# df_filtered['in_top_third_of_percent_range'] = df_filtered.groupby("name").percent_of_births.max()

df_percent_max = df_filtered.groupby("name").percent_of_births.max().reset_index()
df_percent_max = df_percent_max.rename({"percent_of_births": "max_percent_of_births"}, axis='columns')



```

```python
df_filtered = df_filtered.merge(
    df_percent_max,
    on="name"
)
```

```python
df_filtered = df_filtered[
    ((df_filtered.percent_of_births / df_filtered.max_percent_of_births) >= .33)
]
```

```python
df_filtered["in_top_third"] = ((df_filtered.percent_of_births / df_filtered.max_percent_of_births) >= .33)

```

```python
APP_NAME = "tester15"



# Build App
app = JupyterDash(__name__)
app.layout = html.Div([
    html.H1("Baby Names"),
    dcc.Graph(id=APP_NAME),
    html.Label([
        "year",
        dcc.Dropdown(
            id='year-dropdown', clearable=False,
            value="2020", options=[
                {'label': year, 'value': year}
                for year in sorted(list(set(df.year)))
            ]
        )
    ]),
    
    html.Label([
        "peak-percent",
        dcc.RangeSlider(
            id='peak-percent-slider',
            min=0,
            max=0.1,
            step=0.001,
            value=[0, 0.1],
            marks={
                0.0: '0.0',
                0.025: '0.025',
                0.05: '0.05',
                0.075: '0.075',
                0.1: '0.1'
            },
        ),
    ]),
    html.Label([
        "first_letter",
        dcc.Dropdown(
            id='first-letter-dropdown', clearable=False,
            value="v", options=[
                {'label': letter, 'value': letter}
                for letter in sorted(list(first_letters))
            ]
        )
    ]),
])

# Define callback to update graph
@app.callback(
    Output(APP_NAME, 'figure'),
    [Input("year-dropdown", "value"),
     Input("first-letter-dropdown", "value"),
     Input("peak-percent-slider", "value")]
)
def update_graph(year_selection, first_letter_selection, peak_percent_selection):
    names_to_show = set(df_filtered[(df_filtered["year"] == year_selection) 
                                    & (df_filtered["in_top_third"])
                                    & (df_filtered["name"].str.startswith(first_letter_selection))
                                    & (df_filtered["max_percent_of_births"] >= peak_percent_selection[0])
                                    & (df_filtered["max_percent_of_births"] <= peak_percent_selection[1])].name)
    
    traces = list()
    
    #todo: fix this hacky way to make the x-axis linear
    traces.append(go.Scatter(x=sorted(list(set(df_filtered.year))),
                             y=sorted(list(set(df_filtered.percent_of_births))),
                             name=""))
    
    for name in names_to_show:
        name_df = df[(df["name"] == name)]
        traces.append(go.Scatter(x=name_df.year,
                                 y=name_df.percent_of_births,
                                 name=name))
                                 
    
    #display_df = df[df["name"].isin(names)]
    
    return go.Figure(
        data=traces
    )

# Run app and display result inline in the notebook
app.run_server(mode='external')
```
